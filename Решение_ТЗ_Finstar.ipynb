# Задание №1 
## SQL

Дано 2 таблицы:
	1) employee - справочник сотрудников
	2) tasks - задачи, назначенные на сотрудников
необходимо написать запрос, который покажет всех сотрудников, у кого в работе менее трех задач
Результат предоставить в виде: имя сотрудника, кол-во задач в работе.


## Решение:

SELECT e.emp_name, COUNT(t.id) AS task_count

FROM #employee e

LEFT JOIN #tasks t ON t.assignee_id = e.id

GROUP BY e.id, e.emp_name

HAVING COUNT(t.id) < 3;

# Задание №2 
## SQL

Дано 2 таблицы:
	1) payments_principal - платежи по основному долгу
	2) payments_interest - платежи по процентам
Необходимо написать запрос, который покажет сумму платежей по основному долгу и процентам. 
Суммы необходимо считать только по тем договорам, которые платили и основной долг и проценты.
Результат предоставить в виде: id договора, сумма платежей по основному долгу, сумма платежей по процентам.


## Решение:

SELECT p_princ.agr_id,
    
    SUM(p_princ.payment_sum) AS sum_of_princ_p,
    
    SUM(p_int.payment_sum) AS sum_of_int_p
    
FROM #payments_principal AS p_princ
    
INNER JOIN #payments_interest AS p_int ON p_princ.agr_id = p_int.agr_id
    
WHERE p_princ.payment_sum != 0 AND p_int.payment_sum != 0

GROUP BY p_princ.agr_id;

# Задание №3 
## Конверсия

Магазин ежедневно посещает 8000 посетителей, из которых 880 становится покупателями. Выручка за день при этом составляет 2 001 000 рублей при стоимости одной позиции в чеке 890 рублей. Во время сезонной распродажи трафик увеличился на 20%, а стоимость одной позиции в чеке наоборот уменьшилась на 20%.
Рассчитайте конверсию, чтобы сохранить ежедневную выручку в прежнем объеме при среднем количестве позиций в чеке 3,129.


## Решение:

```python
X = 8000
Z = 880
C = 2001000
Y = 890
AVER_TIC = 3.129
X_20 = 0.20
Y_20 = 0.20

# Решение
X_new = X + (X * X_20)
Y_new = Y - (Y * Y_20)
Z_neces = C / (AVER_TIC * Y_new)
Conversion = (Z_neces / X_new) * 100

print(f"Чтобы сохранить выручку в прежнем объеме, нам нужно иметь примерно '{Z_neces:.0f}' покупателей в день") 
print(f"Это составляет примерно '{Conversion:.2f}%' конверсии")
```

## Ответ: ≈9.36%

Во время сезонной распродажи трафик увеличился на 20%, то есть новое количество посетителей: X + (X * 0.20) | 9600 посетителей

А стоимость одной позиции в чеке наоборот уменьшилась на 20%: Y - (Y * 0.20) | 712 рублей

ежедневные посетители - X, 
окнчательные покупатели - Z, 
выручка за день - C, 
стоимость одной позиции в чеке - Y, 
среднее количество позиций в чеке - AVER_TIC, 
увеличение трафика - X_20, 
уменьшение стоимости чека - Y_20, 
необходимые покупатели - Z_neces.

Чтобы сохранить ежедневную выручку в прежнем объеме при среднем количестве позиций в чеке 3129, конверсия должна быть ровна 9.36%

## График

На графике ниже в столбцах показано количество клиентов, которые закрыли заем в конкретный день.
Линиями показан Retention X days = количество клиентов, которые взяли новый заем (доля из тех, кто закрыл – в зеленых столбцах) в течение X дней после даты закрытия займа.
Например, на дату Х закрылось 4000 договоров. Из них (голубая линия) сразу же в дату закрытия 30% клиентов взяли новый заем. По результатам двух дней (оранжевая линия) взяли уже 60% (т.е. 30% в тот же день и еще 30% на следующий). На графике логика по поколениям закрытых договоров и линии привязаны к каждому конкретному поколению, идут накопительным итогом.
Какие выводы можно сделать на основе этого графика?


## Ответ:

В день закрытия займа (показатель голубой линии) значительная часть клиентов сразу же берет новый заем. Например, на дату 2-Sep было закрыто 4200 договоров, и 20% из них сразу же оформили новый заем, что указывает на высокую потребность клиентов в финансировании.

Оранжевая линия, отображающая процент клиентов, взявших новый заем в течение одного дня после закрытия займа, показывает скорость возврата клиентов за новыми займами и составляет 30%. Если через два дня этот процент увеличивается до 40%, это означает, что еще 10% (40%-30%=10%) клиентов решили взять новый заем на третий день после закрытия. (на графике это показано желтым закрашенным квадратиком)




Если процент на желтой линии высок, это указывает на хорошее удержание клиентов в течение месяца после закрытия займа. Анализ желтой линии в сочетании с остальными линиями помогает оценить общую эффективность удержания клиентов и выявить, как быстро они возвращаются за новыми займами после закрытия предыдущих.

Для клиентов, которые закрыли займы после 16-Sep, данные не могут быть полностью проанализированы для 30-дневного периода, поскольку с момента их закрытия займов не прошло 30 дней. Поэтому желтая линия останавливается на 16-Sep, поскольку это последний день, для которого можно полностью измерить 30-дневный ретеншн.

4-Sep, 11-Sep, 18-Sep - синяя линия падает до почти 0%, что указывает на практически полное отсутствие новых займов в день закрытия. Это может быть связано с событиями, например, выходными днями. (на графике это обозначено зелеными закрашенными квадратиками)
